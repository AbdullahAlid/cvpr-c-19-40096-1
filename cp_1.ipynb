```python
import matplotlib.pyplot as plt
import pandas as pd
import random
import math
```

```python
df = pd.read_csv('diagonal.csv')
x1 = df['X1'].values.tolist()
x2 = df['X2'].values.tolist()
y = df['Y'].values.tolist()

for i in range(len(y)):
    if y[i] == 0:
        plt.plot(x1[i],x2[i],'r+')
    else:
        plt.plot(x1[i],x2[i],'g+')
```
